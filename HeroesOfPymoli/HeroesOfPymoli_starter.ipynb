```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_df = pd.read_csv(file_to_load)
purchase_df.head()
```

### Changing Column Names

```python
# Renaming columns to get rid of spaces
purchase_renamed_df = purchase_df.rename(columns = {'Purchase ID' : 'purchase_id',
                                                    'Item ID': 'item_id',
                                                    'Item Name' : 'item_name'})
purchase_renamed_df.head()
```

## Player Count

* Display the total number of players


```python
# Counting total number of unique SN values to find the total number of unique players
total_players = len(purchase_renamed_df.SN.unique())
print(f"The total number of players is {total_players}.")
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Counting the number of unique item ID's 
num_unique_items = len(purchase_renamed_df.item_id.unique())

# Calculating the average of the Price series
avg_price = round(purchase_renamed_df.Price.mean(), 2)

# Calculating the number of sales
num_sales = len(purchase_renamed_df.purchase_id)

# Calculating the total revenue
total_revenue = purchase_renamed_df.Price.sum()

# Creating summary data frame
purchasing_analysis_df = pd.DataFrame({'Number_of_Unique_Items': [num_unique_items],
                                       'Average_Price' : [avg_price],
                                       'Number_of_Purchases' : [num_sales], 
                                       'Total_Revenue' : [total_revenue]})

# Reformat displayed data to contain dollar signs
purchasing_analysis_df.Average_Price = purchasing_analysis_df.Average_Price.map("${:.2f}".format)
purchasing_analysis_df.Total_Revenue = purchasing_analysis_df.Total_Revenue.map("${:.2f}".format)


print(f"There are {num_unique_items} unique items.")
print(f"The average purchase price is ${avg_price}.")
print(f"There are {num_sales} purchases total.")
print(f"The total revenue is ${total_revenue}.")

display(purchasing_analysis_df)
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Reformat dataframe to only have unique users and then count by gender
gender_demo = purchase_renamed_df.drop_duplicates('SN').Gender.value_counts()
#print(gender_demo)

male_count = gender_demo[0]
female_count = gender_demo[1]
other_count = gender_demo[2]
gender_counts = [male_count, female_count, other_count]
#print(gender_counts)

# Calculate percentages
male_percent = male_count/total_players
female_percent = female_count/total_players
other_percent = other_count/total_players
gender_percent = [male_percent*100, female_percent*100, other_percent*100]
#print(gender_percent)

genders = ['Male' , 'Female', 'Other']
gender_summary_df = pd.DataFrame({'Gender' : genders, 
                                  'Total_Count' : gender_counts, 
                                  'Percentage_of_Players' : gender_percent})
gender_summary_df = gender_summary_df.set_index('Gender')

gender_summary_df.Percentage_of_Players = gender_summary_df.Percentage_of_Players.map("{0:.2f}%".format)
display(gender_summary_df)
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Assign bin values and names
age_bins = [0, 9, 14, 19, 24, 29, 34, 39, purchase_renamed_df.Age.max()]
age_bins_names = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

# Take dataframe with duplicate users removed to put into bins and calculate age demographics of unique users
no_dupes_df = purchase_renamed_df.drop_duplicates('SN')
no_dupes_df['age_group'] = pd.cut(no_dupes_df.Age, bins = age_bins, labels = age_bins_names, include_lowest = True)
no_dupes_df.head()

# Use groupby to calculate counts and percentages
age_df = no_dupes_df.groupby('age_group')
age_bin_counts = age_df.count()['SN']
age_bin_percent = age_bin_counts/(total_players/100)

# Create summary dataframe
age_demo_df = pd.DataFrame({'Age_Group' : age_bins_names, 
                            'Total_Counts' : age_bin_counts,
                            'Percentage_of_Players' : age_bin_percent})
age_demo_df = age_demo_df.set_index('Age_Group')
age_demo_df.Percentage_of_Players = age_demo_df.Percentage_of_Players.map("{0:.2f}%".format)
display(age_demo_df)
```
